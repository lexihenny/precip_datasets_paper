```python
yrs=np.arange(1979,2023,1)

#ds=xr.open_dataset(dir2+'mswep_threshold_final_98_all.nc')
ds=xr.open_dataset(dir_data+'mswep_threshold_r99p.nc')
threshold_mswep=ds['threshold'].sel(lat=np.arange(-60,70.25,0.25))#.sel(lat=np.arange(-89.75,90,0.25))

#threshold_mswep_mean=threshold_mswep.mean(dim='time',skipna=True)

if 1==1:
    total_list_mswep_past=[]
    freq_list_mswep_past=[]
    intensity_list_mswep_past=[]

    total_list_mswep_past_land=[]
    freq_list_mswep_past_land=[]
    intensity_list_mswep_past_land=[]

    total_list_mswep_nogauge=[]
    freq_list_mswep_nogauge=[]
    intensity_list_mswep_nogauge=[]

    total_list_mswep_nogauge_land=[]
    freq_list_mswep_nogauge_land=[]
    intensity_list_mswep_nogauge_land=[]
    
    total_mswep_past_monthly=np.zeros((12,len(yrs)))
    freq_mswep_past_monthly=np.zeros((12,len(yrs)))
    intensity_mswep_past_monthly=np.zeros((12,len(yrs)))
    
    total_mswep_past_land_monthly=np.zeros((12,len(yrs)))
    freq_mswep_past_land_monthly=np.zeros((12,len(yrs)))
    intensity_mswep_past_land_monthly=np.zeros((12,len(yrs)))
    
    total_mswep_nogauge_monthly=np.zeros((12,len(yrs)))
    freq_mswep_nogauge_monthly=np.zeros((12,len(yrs)))
    intensity_mswep_nogauge_monthly=np.zeros((12,len(yrs)))
    
    total_mswep_nogauge_land_monthly=np.zeros((12,len(yrs)))
    freq_mswep_nogauge_land_monthly=np.zeros((12,len(yrs)))
    intensity_mswep_nogauge_land_monthly=np.zeros((12,len(yrs)))

dir2='/Volumes/Extreme Pro/'
dir3='/Volumes/My Passport/day/'
for i in range(len(yrs)):
    year=yrs[i]
    print(year)
    ds=xr.open_dataset(dir2+'mswep_past_deres_'+str(year)+'.nc')
    precip_mswep_past=ds['precip'].sel(lat=np.arange(-60,70.25,0.25))#.sel(lat=np.arange(-89.75,90,0.25))
    precip_mswep_past=precip_mswep_past.where(precip_mswep_past>=1.)
    ds.close()
    
    if i==0:#land mask option
        ds=xr.open_dataset('/Users/ahenny/Downloads/data.nc')
        t2m=ds['t2m'][0,:,:]
        land_mask=t2m/t2m
        land_mask=land_mask.fillna(0)
        
        lon_east=[x for x in land_mask.longitude.values if x>=180.]
        lon_west=[x for x in land_mask.longitude.values if x<180.]
        l_east=land_mask.sel(longitude=lon_east)
        l_west=land_mask.sel(longitude=lon_west)
        l_east['longitude']=[x-360. for x in lon_east]
        land_mask=xr.concat([l_east,l_west],dim='longitude')
        
        dk=xr.Dataset()
        dk['land']=(('lat','lon'),land_mask.values)
        dk.coords['lat']=land_mask.latitude.values
        dk.coords['lon']=land_mask.longitude.values
        
        land_mask=dk['land']
        land_mask_interp=land_mask.interp_like(precip_mswep_past[0,:,:])
    
    if 1==0:
        threshold_new_mswep=xr.zeros_like(precip_mswep_past)
        for k in range(threshold_new_mswep.time.size):
            date=pd.to_datetime(threshold_new_mswep.time.values[k])
            if date.month==2 and date.day==29:
                threshold_new1=threshold_mswep.sel(time=dt.datetime(1981,2,28))
                threshold_new2=threshold_mswep.sel(time=dt.datetime(1981,3,1))
                threshold_new3=(threshold_new1+threshold_new1)/2.
            else:
                threshold_new3=threshold_mswep.sel(time=dt.datetime(1981,date.month,date.day))

            threshold_new_mswep[k,:,:]=threshold_new3.values
    if 1==1:
        threshold_new_mswep=threshold_mswep
        
    precip_extreme_mswep_past=precip_mswep_past.where(precip_mswep_past>=1.)
    precip_extreme_mswep_past=precip_mswep_past.where(precip_mswep_past>=threshold_new_mswep)
       

    if year<=2020:
        ds=xr.open_dataset(dir2+'mswep_past_deres_nogauge_'+str(year)+'.nc')
        precip_mswep_nogauge=ds['precip'].sel(lat=np.arange(-60,70.25,0.25))#.sel(lat=np.arange(-89.75,90,0.25))
        precip_mswep_nogauge=precip_mswep_nogauge.where(precip_mswep_nogauge>=1.)
        ds.close()
        
    precip_extreme_mswep_past_land=precip_mswep_past.where(land_mask_interp>=0.5)
    #precip_extreme_mswep_past_land=precip_extreme_mswep_past_land.where(precip_extreme_mswep_past_land>=1.)
    precip_extreme_mswep_past_land=precip_extreme_mswep_past_land.where(precip_extreme_mswep_past_land>=threshold_new_mswep)
    
    
    #####MSWEP PAST
    ones=precip_extreme_mswep_past/precip_extreme_mswep_past
    freq=ones.sum(dim='time',skipna=True).fillna(0)
    extreme_sum=precip_extreme_mswep_past.sum(dim='time',skipna=True).fillna(0)
    
    weights=np.cos(np.deg2rad(extreme_sum.lat))
    total_weighted=extreme_sum.weighted(weights)
    freq_weighted=freq.weighted(weights)
    weighted_1=precip_extreme_mswep_past.weighted(weights)
    intensity=weighted_1.mean(skipna=True).values.tolist()
    
    total_list_mswep_past.append(total_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist())
    freq_list_mswep_past.append(freq_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist())
    intensity_list_mswep_past.append(intensity)
    
    for j in range(12):
        dates=[x for x in ones.time.values if pd.to_datetime(x).month==j+1]
        
        field=precip_extreme_mswep_past.sel(time=dates)
        extreme_sum=field.sum(dim='time',skipna=True).fillna(0)
        total_weighted=extreme_sum.weighted(weights)
        total_mean=total_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist()
        total_mswep_past_monthly[j,i]=total_mean
        
        field=ones.sel(time=dates)
        ones_sum=field.sum(dim='time',skipna=True).fillna(0)
        freq_weighted=ones_sum.weighted(weights)
        freq_mean=freq_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist()
        freq_mswep_past_monthly[j,i]=freq_mean
        
        field=precip_extreme_mswep_past.sel(time=dates)
        intensity_weighted=field.weighted(weights)
        intensity_mean=intensity_weighted.mean(skipna=True).values.tolist()
        intensity_mswep_past_monthly[j,i]=intensity_mean
    
    
    precip_mswep_past.close()
    precip_extreme_mswep_past.close()
    ones.close()
    
    ####MSWEP PAST LAND
    ones=precip_extreme_mswep_past_land/precip_extreme_mswep_past_land
    freq=ones.sum(dim='time',skipna=True).fillna(0).where(land_mask_interp>=0.5)
    extreme_sum=precip_extreme_mswep_past_land.sum(dim='time',skipna=True).fillna(0).where(land_mask_interp>=0.5)
    
    weights=np.cos(np.deg2rad(extreme_sum.lat))
    total_weighted=extreme_sum.weighted(weights)
    freq_weighted=freq.weighted(weights)
    weighted_1=precip_extreme_mswep_past_land.weighted(weights)
    intensity=weighted_1.mean(skipna=True).values.tolist()
    
    total_list_mswep_past_land.append(total_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist())
    freq_list_mswep_past_land.append(freq_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist())
    intensity_list_mswep_past_land.append(intensity)

    for j in range(12):
        dates=[x for x in ones.time.values if pd.to_datetime(x).month==j+1]
        
        field=precip_extreme_mswep_past_land.sel(time=dates)
        extreme_sum=field.sum(dim='time',skipna=True).fillna(0).where(land_mask_interp>=0.5)
        total_weighted=extreme_sum.weighted(weights)
        total_mean=total_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist()
        total_mswep_past_land_monthly[j,i]=total_mean
        
        field=ones.sel(time=dates)
        ones_sum=field.sum(dim='time',skipna=True).fillna(0).where(land_mask_interp>=0.5)
        freq_weighted=ones_sum.weighted(weights)
        freq_mean=freq_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist()
        freq_mswep_past_land_monthly[j,i]=freq_mean
        
        field=precip_extreme_mswep_past_land.sel(time=dates).where(land_mask_interp>=0.5)
        intensity_weighted=field.weighted(weights)
        intensity_mean=intensity_weighted.mean(skipna=True).values.tolist()
        intensity_mswep_past_land_monthly[j,i]=intensity_mean
    
    
    precip_extreme_mswep_past_land.close()
    ones.close()
    
    if year<=2020:
        
        ####MSWEP NOGAUGE
        precip_extreme_mswep_nogauge=precip_mswep_nogauge.where(precip_mswep_nogauge>=threshold_new_mswep)
        #precip_extreme_mswep_nogauge=precip_mswep_nogauge.where(precip_mswep_nogauge>=1.)
        
        
        precip_extreme_mswep_nogauge_land=precip_mswep_nogauge.where(land_mask_interp>=0.5)
        #precip_extreme_mswep_nogauge_land=precip_extreme_mswep_nogauge_land.where(precip_extreme_mswep_nogauge_land>=1.)
        precip_extreme_mswep_nogauge_land=precip_extreme_mswep_nogauge_land.where(precip_extreme_mswep_nogauge_land>=threshold_new_mswep)
        
        
        ones=precip_extreme_mswep_nogauge/precip_extreme_mswep_nogauge
        freq=ones.sum(dim='time',skipna=True).fillna(0)
        extreme_sum=precip_extreme_mswep_nogauge.sum(dim='time',skipna=True).fillna(0)
        
        weights=np.cos(np.deg2rad(extreme_sum.lat))
        total_weighted=extreme_sum.weighted(weights)
        freq_weighted=freq.weighted(weights)
        weighted_1=precip_extreme_mswep_nogauge.weighted(weights)
        intensity=weighted_1.mean(skipna=True).values.tolist()
        
        total_list_mswep_nogauge.append(total_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist())
        freq_list_mswep_nogauge.append(freq_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist())
        intensity_list_mswep_nogauge.append(intensity)
        
        
        for j in range(12):
            dates=[x for x in ones.time.values if pd.to_datetime(x).month==j+1]

            field=precip_extreme_mswep_nogauge.sel(time=dates)
            extreme_sum=field.sum(dim='time',skipna=True).fillna(0)
            total_weighted=extreme_sum.weighted(weights)
            total_mean=total_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist()
            total_mswep_nogauge_monthly[j,i]=total_mean

            field=ones.sel(time=dates)
            ones_sum=field.sum(dim='time',skipna=True).fillna(0)
            freq_weighted=ones_sum.weighted(weights)
            freq_mean=freq_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist()
            freq_mswep_nogauge_monthly[j,i]=freq_mean

            field=precip_extreme_mswep_nogauge.sel(time=dates)
            intensity_weighted=field.weighted(weights)
            intensity_mean=intensity_weighted.mean(skipna=True).values.tolist()
            intensity_mswep_nogauge_monthly[j,i]=intensity_mean
            
        precip_extreme_mswep_nogauge.close()
        precip_mswep_nogauge.close()
        ones.close()
        
        ####MSWEP NOGAUGE LAND
        ones=precip_extreme_mswep_nogauge_land/precip_extreme_mswep_nogauge_land
        freq=ones.sum(dim='time',skipna=True).fillna(0).where(land_mask_interp>=0.5)
        extreme_sum=precip_extreme_mswep_nogauge_land.sum(dim='time',skipna=True).fillna(0).where(land_mask_interp>=0.5)
        
        weights=np.cos(np.deg2rad(extreme_sum.lat))
        total_weighted=extreme_sum.weighted(weights)
        freq_weighted=freq.weighted(weights)
        weighted_1=precip_extreme_mswep_nogauge_land.weighted(weights)
        intensity=weighted_1.mean(skipna=True).values.tolist()
        
        total_list_mswep_nogauge_land.append(total_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist())
        freq_list_mswep_nogauge_land.append(freq_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist())
        intensity_list_mswep_nogauge_land.append(intensity)
        
        precip_extreme_mswep_nogauge_land.close()
        ones.close()

        for j in range(12):
            dates=[x for x in ones.time.values if pd.to_datetime(x).month==j+1]

            field=precip_extreme_mswep_nogauge_land.sel(time=dates)
            extreme_sum=field.sum(dim='time',skipna=True).fillna(0).where(land_mask_interp>=0.5)
            total_weighted=extreme_sum.weighted(weights)
            total_mean=total_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist()
            total_mswep_nogauge_land_monthly[j,i]=total_mean

            field=ones.sel(time=dates)
            ones_sum=field.sum(dim='time',skipna=True).fillna(0).where(land_mask_interp>=0.5)
            freq_weighted=ones_sum.weighted(weights)
            freq_mean=freq_weighted.mean(dim=('lat','lon'),skipna=True).values.tolist()
            freq_mswep_nogauge_land_monthly[j,i]=freq_mean

            field=precip_extreme_mswep_nogauge_land.sel(time=dates).where(land_mask_interp>=0.5)
            intensity_weighted=field.weighted(weights)
            intensity_mean=intensity_weighted.mean(skipna=True).values.tolist()
            intensity_mswep_nogauge_land_monthly[j,i]=intensity_mean
```
