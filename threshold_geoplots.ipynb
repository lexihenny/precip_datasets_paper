```python
ds=xr.open_dataset(dir_data+'persiann_threshold_final_98_all.nc')
threshold=ds['threshold']#.sel(lat=np.arange(-89.75,90,0.25))
threshold_mean=threshold.mean(dim='time')
mean=threshold_mean

#ds=xr.open_dataset(dir_data+'persiann_threshold_r95p.nc')
#threshold=ds['threshold']#.sel(lat=np.arange(-89.75,90,0.25))
#mean=threshold

from matplotlib.colors import LinearSegmentedColormap
cmap1=plt.cm.hot_r(np.linspace(0.1,0.5,128))
cmap2=plt.cm.viridis(np.linspace(0.1,1,256))

cmap3=LinearSegmentedColormap.from_list('new_cmap',['orange','m'])
cmap3=cmap3(np.linspace(0,1,256))

colors = np.vstack((cmap2, cmap1,cmap3))
mymap = LinearSegmentedColormap.from_list('my_colormap', colors)

fig=plt.figure(figsize=(21,6.5))
ax=plt.subplot(1,1,1,projection=ccrs.PlateCarree())
cax=ax.contourf(mean.lon.values,mean.lat.values,mean.where(mean>0),levels=np.arange(0,75,5),extend='max',cmap=mymap)
cbar=plt.colorbar(cax,pad=0,fraction=0.046,aspect=16,ticks=[10,20,30,40,50,60,70])

cbar.ax.tick_params(labelsize=24) 
cbar.set_label('mm day$^{-1}$',fontsize=25,labelpad=5)
ax.coastlines(resolution='10m')
countries = cfeature.NaturalEarthFeature(category='cultural',name='admin_0_boundary_lines_land',scale='50m',facecolor='none')
ax.add_feature(countries)

X=5
for j in range(int(mean.lat.size/X)):
    for k in range(int(mean.lon.size/X)):
        if mean[X*j,X*k].values<1.:
            ax.plot(mean.lon.values[X*k],mean.lat.values[X*j],'wo',markersize=1)
            
ax.set_ylim(-60,70)
ax.set_title(r'$\bf{PERSIANN-CDR}$'+' '+r'$\bf{(1991-2020)}$',fontsize=34,pad=5)
plt.show()
```
